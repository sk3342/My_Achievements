The groupby function in Python, particularly in the pandas library, is used for splitting data into groups based on some criteria. It allows you to group data by one or more columns, perform operations on each group, and then combine the results. This is particularly useful for:

Summarizing data

Calculating statistics like mean, sum, count, etc., for each group

Applying custom functions to each group

Aggregating data in various ways

```python
df3.groupby(['race_length','athelete_gender'])['athelete_average_speed'].mean()
```

**Let's check what age groups are best in the 50mi Race (20+ minimum races)**

**The sort_values method in pandas is used to sort a DataFrame or Series by one or more columns. It can sort in ascending or descending order and can handle missing values.**

```python
df3.query('race_length == "50mi"').groupby('athelete_age')['athelete_average_speed'].agg(['mean','count']).sort_values('mean', ascending = False).query('count>19').head(15)
```

```python
# what age group are worst in 50mi race
```

```python
df3.query('race_length == "50mi"').groupby('athelete_age')['athelete_average_speed'].agg(['mean','count']).sort_values('mean', ascending = True).query('count>9').head(20)
```

```python
# seasons for the date  => runners slower in summer than winter/

#spring 3-5
#summer 6-8
#fall 9-11
#winter 12-2

#split between two decimals
```

```python
df3['race_month'] = df3['race_day'].str.split('.').str.get(1).astype(int)
```

```python
df3.head(5)
```

```python
df3['race_season'] = df3['race_month'].apply(lambda x : 'Winter' if x > 11 else 'Fall' if x > 8 else 'Summer' if x>5 else 'Spring' if x>2 else 'Winter')
```

```python
df3.head(55)
```

```python
df3['race_season'].nunique()
```

```python
df3.groupby('race_season')['athelete_average_speed'].agg(['mean','count']).sort_values(by='mean',ascending = False)
```
